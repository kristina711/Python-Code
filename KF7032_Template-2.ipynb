```python
import turtle

def draw_branch(t, branch_length, left_angle, right_angle, reduction_factor, depth):
    if depth == 0:
        return

    # Set color and thickness based on depth
    if depth > 2:
        t.color("brown")
        t.pensize(depth)
    else:
        t.color("green")
        t.pensize(depth)

    # Draw the current branch
    t.forward(branch_length)

    # Draw left branch
    t.left(left_angle)
    draw_branch(t, branch_length * reduction_factor, left_angle, right_angle, reduction_factor, depth - 1)

    # Return to original position and angle
    t.right(left_angle + right_angle)

    # Draw right branch
    draw_branch(t, branch_length * reduction_factor, left_angle, right_angle, reduction_factor, depth - 1)

    # Return to the original position and angle
    t.left(right_angle)
    t.backward(branch_length)

def generate_tree():
    # User input for tree parameters
    left_angle = float(input("Enter the left branch angle (in degrees): "))
    right_angle = float(input("Enter the right branch angle (in degrees): "))
    starting_length = float(input("Enter the starting branch length (in pixels): "))
    reduction_factor = float(input("Enter the branch length reduction factor (e.g., 0.7 for 70%): "))
    depth = int(input("Enter the recursion depth: "))

    # Set up turtle
    screen = turtle.Screen()
    screen.bgcolor("white")
    screen.title("Recursive Tree")

    t = turtle.Turtle()
    t.speed("fastest")
    t.left(90)  # Start pointing upwards
    t.penup()
    t.goto(0, -screen.window_height() // 4)  # Start at the bottom center of the screen
    t.pendown()

    # Draw the tree
    draw_branch(t, starting_length, left_angle, right_angle, reduction_factor, depth)

    # Complete
    screen.mainloop()

# Run the tree generator
generate_tree()

```

```python
def encrypt_text(input_file, output_file, n, m):
    # Read the content of the input file
    with open(input_file, 'r') as file:
        raw_text = file.read()

    # Define encryption logic
    def encrypt_char(c):
        if 'a' <= c <= 'm':  # First half of lowercase letters
            return chr(((ord(c) - ord('a') + n * m) % 26) + ord('a'))
        elif 'n' <= c <= 'z':  # Second half of lowercase letters
            return chr(((ord(c) - ord('a') - (n + m)) % 26) + ord('a'))
        elif 'A' <= c <= 'M':  # First half of uppercase letters
            return chr(((ord(c) - ord('A') - n) % 26) + ord('A'))
        elif 'N' <= c <= 'Z':  # Second half of uppercase letters
            return chr(((ord(c) - ord('A') + (m ** 2)) % 26) + ord('A'))
        else:  # Non-alphabet characters remain unchanged
            return c

    # Encrypt the text
    encrypted_text = ''.join(encrypt_char(c) for c in raw_text)

    # Write the encrypted text to the output file
    with open(output_file, 'w') as file:
        file.write(encrypted_text)

    print(f"Encrypted text saved to '{output_file}'.")


def decrypt_text(input_file, output_file, n, m):
    # Read the content of the input file
    with open(input_file, 'r') as file:
        encrypted_text = file.read()

    # Define decryption logic
    def decrypt_char(c):
        if 'a' <= c <= 'm':  # First half of lowercase letters
            return chr(((ord(c) - ord('a') - n * m) % 26) + ord('a'))
        elif 'n' <= c <= 'z':  # Second half of lowercase letters
            return chr(((ord(c) - ord('a') + (n + m)) % 26) + ord('a'))
        elif 'A' <= c <= 'M':  # First half of uppercase letters
            return chr(((ord(c) - ord('A') + n) % 26) + ord('A'))
        elif 'N' <= c <= 'Z':  # Second half of uppercase letters
            return chr(((ord(c) - ord('A') - (m ** 2)) % 26) + ord('A'))
        else:  # Non-alphabet characters remain unchanged
            return c

    # Decrypt the text
    decrypted_text = ''.join(decrypt_char(c) for c in encrypted_text)

    # Write the decrypted text to the output file
    with open(output_file, 'w') as file:
        file.write(decrypted_text)

    print(f"Decrypted text saved to '{output_file}'.")


def check_correctness(original_file, decrypted_file):
    # Read the content of the original and decrypted files
    with open(original_file, 'r') as file:
        original_text = file.read()

    with open(decrypted_file, 'r') as file:
        decrypted_text = file.read()

    # Compare the texts
    return original_text == decrypted_text


# Example Usage
n, m = 3, 2  # Sample user inputs
input_file = "/Users/namansisodia/Downloads/raw_text.txt"  # Full path to the raw text file
encrypted_file = "/Users/namansisodia/Downloads/encrypted_text.txt"  # Output file for encrypted text
decrypted_file = "/Users/namansisodia/Downloads/decrypted_text.txt"  # Output file for decrypted text

# Run the functions
encrypt_text(input_file, encrypted_file, n, m)
decrypt_text(encrypted_file, decrypted_file, n, m)

# Check correctness of the decryption
if check_correctness(input_file, decrypted_file):
    print("The decryption was successful and matches the original text.")
else:
    print("The decryption failed.")
```

```python
import os
import pandas as pd

def calculate_average_temperature(input_folder, output_file):
    """
    Calculates the average temperature for each month across all CSV files.
    """
    all_data = []
    
    # Read all CSV files in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".csv"):
            file_path = os.path.join(input_folder, file_name)
            data = pd.read_csv(file_path)
            all_data.append(data)

    # Combine all data into a single DataFrame
    combined_data = pd.concat(all_data, ignore_index=True)
    
    # Calculate average temperature for each month
    month_columns = [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ]
    monthly_averages = combined_data[month_columns].mean()

    # Save to file
    with open(output_file, "w") as file:
        file.write("Average Monthly Temperatures (°C):\n")
        for month, avg_temp in monthly_averages.items():
            file.write(f"{month}: {avg_temp:.2f}\n")
    
    print(f"Average temperatures saved to '{output_file}'.")

def find_largest_temperature_range(input_folder, output_file):
    """
    Finds the station with the largest temperature range.
    """
    all_data = []
    
    # Read all CSV files in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".csv"):
            file_path = os.path.join(input_folder, file_name)
            data = pd.read_csv(file_path)
            all_data.append(data)

    # Combine all data into a single DataFrame
    combined_data = pd.concat(all_data, ignore_index=True)
    
    # Calculate the temperature range for each station
    combined_data["Temperature_Range"] = combined_data.loc[:, "January":"December"].max(axis=1) - combined_data.loc[:, "January":"December"].min(axis=1)
    max_range = combined_data["Temperature_Range"].max()
    largest_range_stations = combined_data[combined_data["Temperature_Range"] == max_range]
    
    # Save to file
    with open(output_file, "w") as file:
        file.write("Station(s) with the Largest Temperature Range:\n")
        for _, row in largest_range_stations.iterrows():
            file.write(f"{row['STATION_NAME']} (Range: {max_range:.2f}°C)\n")
    
    print(f"Largest temperature range station(s) saved to '{output_file}'.")

def find_warmest_and_coolest_stations(input_folder, output_file):
    """
    Finds the warmest and coolest stations based on average annual temperature.
    """
    all_data = []
    
    # Read all CSV files in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".csv"):
            file_path = os.path.join(input_folder, file_name)
            data = pd.read_csv(file_path)
            all_data.append(data)

    # Combine all data into a single DataFrame
    combined_data = pd.concat(all_data, ignore_index=True)
    
    # Calculate the average annual temperature for each station
    combined_data["Average_Annual_Temperature"] = combined_data.loc[:, "January":"December"].mean(axis=1)
    max_avg_temp = combined_data["Average_Annual_Temperature"].max()
    min_avg_temp = combined_data["Average_Annual_Temperature"].min()
    warmest_stations = combined_data[combined_data["Average_Annual_Temperature"] == max_avg_temp]
    coolest_stations = combined_data[combined_data["Average_Annual_Temperature"] == min_avg_temp]
    
    # Save to file
    with open(output_file, "w") as file:
        file.write("Warmest Station(s):\n")
        for _, row in warmest_stations.iterrows():
            file.write(f"{row['STATION_NAME']} (Average: {max_avg_temp:.2f}°C)\n")
        file.write("\nCoolest Station(s):\n")
        for _, row in coolest_stations.iterrows():
            file.write(f"{row['STATION_NAME']} (Average: {min_avg_temp:.2f}°C)\n")
    
    print(f"Warmest and coolest stations saved to '{output_file}'.")

# Main Program
input_folder = "/Users/namansisodia/Downloads/HIT137 Assignment 2 SS 2024 (1)/temperature_data"
average_temp_file = "/Users/namansisodia/Downloads/average_temp.txt"
largest_range_file = "/Users/namansisodia/Downloads/largest_temp_range_station.txt"
warmest_coolest_file = "/Users/namansisodia/Downloads/warmest_and_coolest_station.txt"

# Call functions
calculate_average_temperature(input_folder, average_temp_file)
find_largest_temperature_range(input_folder, largest_range_file)
find_warmest_and_coolest_stations(input_folder, warmest_coolest_file)
```

```python
# Task 2: Weather Forecast App
# Import necessary modules
import requests

def fetch_weather(city_name):
    """
    Function to fetch weather data for a given city from an API.
    Args:
        city_name (str): Name of the city to fetch the weather for.
    Returns:
        dict: Parsed weather data if the API call is successful, or None if it fails.
    """
    # Define the API URL and key (replace 'YOUR_API_KEY' with your actual API key)
    api_key = "YOUR_API_KEY"
    base_url = "http://api.openweathermap.org/data/2.5/weather"

    # Create the complete API request URL
    complete_url = f"{base_url}?q={city_name}&appid={api_key}&units=metric"

    # Make the API request
    response = requests.get(complete_url)

    if response.status_code == 200:
        # Parse the JSON data from the API response
        data = response.json()

        # Extract relevant weather data
        weather_data = {
            "city": data["name"],
            "temperature": data["main"]["temp"],
            "humidity": data["main"]["humidity"],
            "description": data["weather"][0]["description"],
        }
        return weather_data
    else:
        # Handle the error if the city is not found or there is another issue
        print(f"Error: Unable to fetch weather data for {city_name}. (Status Code: {response.status_code})")
        return None

def display_weather(weather_data):
    """
    Function to display weather information in a formatted way.
    Args:
        weather_data (dict): Weather data to display.
    """
    print("\nWeather Information:")
    print(f"City: {weather_data['city']}")
    print(f"Temperature: {weather_data['temperature']} °C")
    print(f"Humidity: {weather_data['humidity']}%")
    print(f"Description: {weather_data['description'].capitalize()}")

# Main application logic
def main():
    """
    Main function to run the weather forecast app.
    """
    # List to store weather data for multiple cities
    all_weather_data = []

    # Input: Get cities from the user
    print("Welcome to the Weather Forecast App!")
    cities = input("Enter city names separated by commas: ").split(",")

    # Strip extra whitespace and format city names
    cities = [city.strip().title() for city in cities]

    # Fetch and display weather data for each city
    for city in cities:
        weather = fetch_weather(city)
        if weather:
            all_weather_data.append(weather)
            display_weather(weather)

    # Optionally, display all data at once
    print("\nSummary of Weather Data:")
    for weather in all_weather_data:
        print(weather)

# Run the app
if __name__ == "__main__":
    main()
```
